```python
df_tracks.set_index('release_date', inplace = True)
df_tracks.index  = pd.to_datetime(df_tracks.index)
df_tracks.head()
```

```python
df_tracks['duration'] = df_tracks['duration_ms'].apply(lambda x: round(x/1000))
df_tracks.drop('duration_ms', inplace = True, axis = 1)
```

```python
df_tracks.duration.head()
```

```python
df_sample = df_tracks.sample(int(0.004*len(df_tracks)))
print(len(df_sample))
```

```python
plt.figure(figsize=(12,6))
sns.regplot(data = df_sample, y ='loudness',x='energy', color = 'c' ).set(title ='Loudness vs Energy')
```

```python
plt.figure(figsize=(12,6))
sns.regplot(data = df_sample, y ='popularity',x='acousticness', color = 'b' ).set(title ='popularity vs acousticness')
```

```python
df_tracks['dates']=df_tracks.index.get_level_values('release_date')
df_tracks.dates=pd.to_datetime(df_tracks.dates)
years=df_tracks.dates.dt.year
```

```python
sns.displot(years,discrete=True,aspect=2,height=4,kind='hist').set(title='Number songs per year')
```

```python
# Heatmap

corr_df = df_tracks.drop(['key', 'mode', 'explicit'], axis = 1).corr(method="pearson")
plt.figure(figsize=(14,6))
heatmap = sns.heatmap(corr_df,annot =True,fmt ='.1g',vmin =-1,vmax =1,center=0, cmap='inferno', linewidhts =1, linecolor ='Black')
heatmap.set_title("Correaltion Heatmap between variable")
heatmap.set_xticklables(heatmap.get_xticklables(), rotation = 90)
```

```python
# Line charts

total_dr=df_tracks.duration
sns.set_style(style="whitegrid")
fig_dims = (10,5)
fig, ax = plt.subplots(fig_size=fig_dims)
fig = sns.lineplot(x=years, y = total_dr, ax=ax).set(title="Year vs Duration")
plt.xticks(rotation=60
```

```python
# Bar charts

total_dr = df_tracks.duration
fig_dims = (18,7)
fig, ax = plt.subplots(figsize = fig_dims)
fig = sns.barplot(x= years, y=total_dr,ax=ax,errwidht=False).set(title="Years vs Duration")
plt.xticks(rotation=90)
```

```python
df_genre = pd.read_csv("SpotifyFeatures.csv")
df_genre.head()
```

```python
plt.title("Duration of the songs in Different Genres")
sns.color_palette("rocket", as_cmap = True)
sns.barplot(y='genre', x ='duration_ms', data = df_genre)
plt.xlabel("Duration in milli seconds")
plt.ylabel("Genres")
```

```python
sns.set_style(style="darkgrid")
plt.figure(figsize=(10,5))
famous = df_genre.sort_values("popularity", ascending = False).head(10)
sns.barplot(y="genre", x="popularity", data = famous).set(title="Top 5 Genres by Popularity")
```
